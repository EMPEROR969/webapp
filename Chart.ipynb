```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

LINE CHART

```python
#kode dibawah ini merupakan array dari nilai
data = np.array([90,89,87,85,95])

# plt.plot(data, linestyle = 'dotted')
# plt.plot(data, color = 'red')

#kode dibawah ini untuk membuat jenis grafik nya 
plt.plot(data, linestyle = 'dashed')

#kode ini untuk menampilkan
plt.show()
```

```python
data = np.array([70,80,85,90,100])
plt.plot(data, linestyle = 'dotted')
plt.show()
```

```python
data = np.array([80,85,70,60,90])
plt.plot(data, color = 'navy')
plt.show()
```

```python
data1 = np.array(['A','B','C','D','E'])
data2 = np.array([90,80,70,60,50])
plt.plot(data1, data2)
plt.show()
```

BAR CHART

```python
data1 = np.array(["A","B","C","D","E"])
data2 = np.array([50,60,70,80,90])
plt.bar(data1, data2)
plt.show()
```

```python
data1 = np.array(["A","B","C","D","E"])
data2 = np.array([50,60,70,80,90])
plt.barh(data1, data2)
plt.show()
```

```python
data1 = np.array(['Steven', 'Kevin', 'Theo', 'Erick', 'Vino'])
data2 = np.array([70,80,85,90,60])
plt.bar(data1,data2,color = 'Orange')
plt.show()
```

```python
data1 = np.array(['Steven', 'William', 'Mama', 'Papa', 'Om'])
data2 = np.array([70,80,85,90,60])
plt.barh(data1,data2,color = 'Orange', height = 0.1)
plt.show()
```

```python
data1 = np.array(['Steven', 'William', 'Mama', 'Papa', 'Om'])
data2 = np.array([70,80,85,90,60])

#Width hanya bisa di bar. height bisa di barh atau horizontal. 
plt.bar(data1,data2,color = 'Orange', width = 0.1)
plt.show()
```

PIE CHART

```python
data = np.array([60,65,70,75,80])
label = ['apple', 'banana', 'orange', 'mango', 'grape']
plt.pie(data, labels = label)
plt.show()
```

```python
data = np.array([70,80,60,50,90])
label = ['Kak Richie', 'Kak Lia', 'Kak Rere', 'Kak Alvin', 'Kak Laras']
myexplode = [0.2,0,0,0.1,0]
color = ['red', 'yellow','black','purple','pink']
plt.pie(data, labels = label, explode = myexplode, colors = color)
plt.show()
```
